# Environment Setup

```python
# Import the packages
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
# Import __future__.division to force all division results to float
from __future__ import division
```

```python
# Read the data
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

# Basic Plot

```python
# RGB colours
cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_grey = (171/255, 171/255, 171/255)
# Lists of degrees by category
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences',
             'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism',
                 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education',
              'Agriculture','Business', 'Architecture']

# Generate a figure instance
fig = plt.figure(figsize=(16,16))

# First column: STEM degrees
# Loop from 0 t0 5
for sp in range(0,6):
    # Add subplot at positions 1, 4, 7, 10, 13 and 16
    ax = fig.add_subplot(6,3,sp*3+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    # Add text annotations to topmost and bottommost plots
    if sp == 0:
        ax.text(2002, 87, 'Women')
        ax.text(2005, 8, 'Men')
    elif sp == 5:
        ax.text(2005, 90, 'Men')
        ax.text(2001, 0, 'Women')
      
# Second column: liberal arts degrees
for sp in range(0,5):
    ax = fig.add_subplot(6,3,sp*3+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[sp])
    # Add text annotation to topmost plot
    if sp == 0:
        ax.text(2002, 79, 'Women')
        ax.text(2005, 14, 'Men')
        
# Third column: other degrees
for sp in range(0,6):
    ax = fig.add_subplot(6,3,sp*3+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[sp])
    # Add text annotations to topmost and bottommost plots
    if sp == 0:
        ax.text(2002, 94, 'Women')
        ax.text(2005, 0, 'Men')
    elif sp == 5:
        ax.text(2005, 64, 'Men')
        ax.text(2001, 23, 'Women')
    
# Save the figure
plt.savefig("gender_degrees.png")
plt.show()
```

# Removing spines and ticks
